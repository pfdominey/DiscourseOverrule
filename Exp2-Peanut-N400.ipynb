```python
import numpy as np
from wikipedia2vec import Wikipedia2Vec

import matplotlib.pyplot as plt

import scipy
from scipy import stats
```

```python
data = np.load('Exp2C-reservoir-data.npy')
data = data.T
```

```python
# convert semantic relatedness to N400:  N400 = (1 - semanticrelatedness)
for i in range(len(data)):
    for j in range(len(data[0])):
        data[i][j] = 1 - data[i][j]
```

```python
# Create a figure instance
#fig = plt.figure(1, figsize=(9, 6))
#fig = plt.figure(1)
fig= plt.figure(figsize=(6,4))
# Create an axes instance
ax = fig.add_subplot(111)

# Create the boxplot
bp = ax.boxplot(data)

ax.set_xticklabels(['Love Before', 'Salted Before', 'Love After', 'Salted After'])
#plt.title('Before Discourse')
#plt.xlabel('Semantic Relatedness')
plt.ylabel('Semantic Relatedness')
fig.tight_layout()
plt.savefig('./peanut-results/Before-after-50res-N400.png')
plt.show()
```

```python
fig= plt.figure(figsize=(12,4))
# Cut your window in 1 row and 2 columns, and start a plot in the first part
plt.subplot(131)
plt.hist(data[0], 8, alpha=0.5, label='Love')
plt.hist(data[1], 8, alpha=0.5, label='Salted')

# Add title and axis names
plt.title('Before discourse')
plt.xlabel('Predicted N400')
plt.ylabel('Instances')
plt.legend(loc='best')
plt.xlim(0.30, 0.60)
plt.ylim(0, 18)
 
# And now add something in the second part:
plt.subplot(132)
plt.hist(data[:,2], 8, alpha=0.5, label='Love')
plt.hist(data[:,3], 8, alpha=0.5, label='Salted')


# Add title and axis names
plt.title('After discourse')
plt.xlabel('Predicted N400')
plt.ylabel('Instances')
plt.legend(loc='best')

plt.xlim(0.30, 0.60)
plt.ylim(0, 18)

# And now add something in the THIRD part:
# Create an axes instance
ax = fig.add_subplot(133)

# Create the boxplot
bp = ax.boxplot(data)

ax.set_xticklabels(['Love-B', 'Salted-B', 'Love-A', 'Salted-A'])
plt.title('Before      After')
plt.xlabel('Target words before vs after discourse')
plt.ylabel('Predicted N400')

fig.tight_layout()
 
# Show the graph
plt.savefig('Peanut-50-Res-Hist-N400-R1.png', dpi=1200)
plt.show()
```
