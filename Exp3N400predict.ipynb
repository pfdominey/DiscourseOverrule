```python
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import pyplot
import nltk

import csv
import operator
import scipy
from scipy import stats
```

```python
#load the data generated in Exp3A-metusalem2012_average 
data = np.load("Exp3A-average-data.npy")
```

```python
# get into the right format, and perform the N400 = 1 - semantic relatedness transformation
data = data.T
print(len(data[0]))
print(len(data))
for i in range(len(data)):
    for j in range(len(data[0])):
        data[i][j] = 1 - data[i][j]
```

```python
# graphic for average vector test 


fig= plt.figure(figsize=(18,6))
plt.title('Average Vector')

# Cut your window in 1 row and 2 columns, and start a plot in the first part
plt.subplot(131)
plt.hist(data[:,0], 8, alpha=0.5, label='Expected')
plt.hist(data[:,1], 8, alpha=0.5, label='Related')
plt.hist(data[:,2], 8, alpha=0.5, label='Unrelated')

# Add title and axis names
plt.title('Sentence context')
plt.xlabel('Predicted N400')
plt.ylabel('Instances')
plt.legend(loc='best')

plt.xlim(0.2, 0.75)
plt.ylim(0, 30)
 
# And now add something in the second part:
plt.subplot(132)
plt.hist(data[:,3], 8, alpha=0.5, label='Expected')
plt.hist(data[:,4], 8, alpha=0.5, label='Related')
plt.hist(data[:,5], 8, alpha=0.5, label='Unrelated')

# Add title and axis names
plt.title('Discourse context')
plt.xlabel('Predicted N400')
plt.ylabel('Instances')
plt.legend(loc='best')

plt.xlim(0.2, 0.75)
plt.ylim(0, 30)
 
 
# now 

ax = fig.add_subplot(133)


# Create the boxplot
bp = ax.boxplot(data, patch_artist=True)

colors = ['lightblue', 'moccasin', 'lightgreen', 'lightblue', 'moccasin', 'palegreen']
 
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
ax.set_xticklabels(['Expected', 'Related', 'Unrelated', 'Expected', 'Related', 'Unrelated'])
# Add title and axis names
plt.title('Sentence            Discourse')
#plt.xlabel('Target words in Sentence vs Discourse')
plt.ylabel('Predicted N400')
#fig.tight_layout()

plt.savefig('Figure5.png', dpi=1200)
```
